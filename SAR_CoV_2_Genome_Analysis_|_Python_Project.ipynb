The following is the complete *reference* genome of SAR-CoV-2, NCBI accession [NC_045512](https://www.ncbi.nlm.nih.gov/nuccore/NC_045512)

```
ncov = ""
```

**Q1.** What is the genome size of SAR-CoV-2?

```
size = len(ncov)
print ("Genome size = ", size)
```

**Q2.** What is the [GC-content](https://en.wikipedia.org/wiki/GC-content) of the virus genome?

```
g = ncov.count("G")
c = ncov.count("C")
print ("GC-content = ", 100*(g+c)/size)
```

**Q3.** What is the [dinucleotide](https://www.biologyonline.com/dictionary/dinucleotide) frequency in the genome SAR-CoV-2?

```
dinucleotides = []
for nt1 in ['A', 'C', 'T', 'G']:
  for nt2 in ['A', 'C', 'T', 'G']:
    print (nt1, nt2)
    dinucleotides.append(nt1 + nt2)
print(dinucleotides)
```

```
for dinucleotide in dinucleotides:
  count = ncov.count(dinucleotide)
  print (dinucleotide, count)
```
